This Notebook shows how the confidence levels in the power spectra plots shown in Figure of the paper are calculated.

```python
import numpy as np

import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse
import matplotlib.image as mpimg
import matplotlib as mpl
mpl.rc('font',family='serif')

from astropy.stats import LombScargle
from astropy.modeling import models, fitting, custom_model
from astropy.table import Table
from astropy import constants as const
from astropy import units as u

from scipy.interpolate import interp1d

import corner

from stingray.simulator import simulator

from python_utils import PowerSpec
```

```python
# convenience ephemeris function
def orb_ephemeris(T):
    return (T - 2455912.83987)/(0.3667200) - (T - 2455912.83987)//(0.3667200)

# A convenience function to convert data from mag to flux
def flux_calibrate(m, filt, m_err=[]):
    SDSS_calibs = {'u': [3767, 1.4e-10], 'g': [3631, 0.9e-10], 'r': [3631, 1.2e-10], 'i': [3631, 1.8e-10], 'z': [3565, 7.4e-10]}#https://ned.ipac.caltech.edu/help/sdss/dr6/photometry.html#asinh
    f0 = SDSS_calibs[filt][0]
    b = SDSS_calibs[filt][1]
    flux = 2*b*f0*np.sinh(m/(-2.5/np.log(10))-np.log(b))
    
    flux = 1000*flux # Turning into mJy
    if len(m_err) != 0:
        err_flux = np.sqrt((m_err**2)*((-1.842*b*f0*np.cosh(np.log(b)+0.921034*m))**2))
        err_flux = 1000*err_flux # Turning into mJy
        return flux, err_flux
    else:
        return flux

# A convenience function to deredden the data
def deredden(M,E_bv,sdss_filt):
    SDSS_extincts = {'u': 5.155, 'g': 3.793, 'r': 2.751, 'i': 2.086, 'z': 1.479} # http://iopscience.iop.org/article/10.1086/324741/pdf Table 22
    A_mag = E_bv*SDSS_extincts[sdss_filt]
    return M - A_mag
```

The following is a convence function which takes time, flux, error, a plotting axis, a minimum period, a maximum period, a bin width, and an upper limit function. It calculates the power spectrum for time series between pmin and pmax. It also bins this power spectrum and fits a broken power law to the power spectrum. The power spectrum, broken power law fit, and upper limits (if passed) are then plotted on axis. Note that the function PowerSpec is a convenience function located in my python_utils packge

```python
def fit_PS(time,flux,error,ax, pmin, pmax, bin_width, ul=None):
    oversampling=10
    ps = PowerSpec(time,flux,error)
    ps.runspec(pmin=pmin, pmax=pmax, norm='psd', oversampling=10)

    broken_pl_init = models.BrokenPowerLaw1D(amplitude=3.51686205e+01,x_break=1.40269882e-03,alpha_1=0.3564771551659097,alpha_2=1.96878805e+00)
    broken_pl_init.x_break.fixed = True
    broken_pl_init.alpha_1.fixed = True
    pl_init = models.PowerLaw1D(amplitude=1,alpha=1.25)
    fit_pl = fitting.LevMarLSQFitter()

    data_table = Table([ps.freq_range,ps.PS], names=('Freq', 'PS'), meta={'name': 'flux_table'})
    freq_bin = np.trunc(np.array(data_table['Freq']/(24*60*60)) / bin_width)
    data_table_grouped = data_table.group_by(freq_bin)
    data_table_binned = data_table_grouped.groups.aggregate(np.median)
    data_table_rms = data_table_grouped.groups.aggregate(np.std)

    for i,j in enumerate(np.unique(freq_bin)):
        no_points = len(freq_bin[freq_bin==j])
        data_table_rms['PS'][i] = data_table_rms['PS'][i]/np.sqrt(no_points/oversampling)

    broken_pl = fit_pl(broken_pl_init, data_table_binned['Freq']/(24*60*60),data_table_binned['PS'], weights=1/data_table_rms['PS'])
    pl = fit_pl(pl_init, data_table_binned['Freq']/(24*60*60),data_table_binned['PS'], weights=1/data_table_rms['PS'])

    ax.plot(ps.freq_range/(24*60*60),ps.PS,color='grey',alpha=0.4,label='PS')
    ax.errorbar(data_table_binned['Freq']/(24*60*60),data_table_binned['PS'],xerr=bin_width/2,yerr=data_table_rms['PS'],fmt='k.',label='Binned PS')
    ax.plot(ps.freq_range/(24*60*60),broken_pl(ps.freq_range/(24*60*60)),color='C0',linewidth=4,label='Broken Powerlaw')
    if ul!=None:
        ax.plot(ps.freq_range/(24*60*60),ul(ps.freq_range/(24*60*60)),color='C3',label=r'3$\sigma$')
    ax.set_xlim(3e-6,max(ps.freq_range/(24*60*60)))
    ax.set_yscale("log")
    ax.set_xscale("log")
    ax.legend(loc=3)

    logL = -0.5*np.sum(((data_table_binned['PS']-broken_pl(data_table_binned['Freq']/(24*60*60)))/(data_table_rms['PS']))**2)
    #print (logL)
    
    return ps, broken_pl
```

## Loading in UCAM data

```python
data = np.genfromtxt("lc.dat")

time_BJD = data[:,0]
phase = data[:,1]
i_mag = data[:,2]
i_err = data[:,3]
g_mag = data[:,4]
g_err = data[:,5]
u_mag = data[:,6]
u_err = data[:,7]

time_u_BJD = time_BJD[~np.isnan(u_mag)]
phase_u = phase[~np.isnan(u_mag)]
u_err = u_err[~np.isnan(u_mag)]
u_mag = u_mag[~np.isnan(u_mag)]

i_mag = deredden(i_mag,0.0374,'i')
g_mag = deredden(g_mag,0.0374,'g')
u_mag = deredden(u_mag,0.0374,'u')
```

## Loading in best fit model from ELC

```python
model_u_mag = np.genfromtxt("modelU.mag")
model_g_mag = np.genfromtxt("modelB.mag")
model_i_mag = np.genfromtxt("modelR.mag")

model_u_flux = flux_calibrate(model_u_mag[:,1],'u')
model_g_flux = flux_calibrate(model_g_mag[:,1],'g')
model_i_flux = flux_calibrate(model_i_mag[:,1],'i')

#Interpolating the best fit models so they can be plotted at same time resolution as the data
model_func_u_mag = interp1d(model_u_mag[:,0],model_u_mag[:,1], kind='cubic')
model_func_g_mag = interp1d(model_g_mag[:,0],model_g_mag[:,1], kind='cubic')
model_func_i_mag = interp1d(model_i_mag[:,0],model_i_mag[:,1], kind='cubic')

model_func_u_flux = interp1d(model_u_mag[:,0],model_u_flux, kind='cubic')
model_func_g_flux = interp1d(model_g_mag[:,0],model_g_flux, kind='cubic')
model_func_i_flux = interp1d(model_i_mag[:,0],model_i_flux, kind='cubic')
```

## Lomb Scargle Periodogram

The following shows how the 3 sigma limits are derived. The example uses the i band data from 2017-10-16. We start by first fitting a broken PL to the power spectrum

```python

oversampling=10
ps = PowerSpec(i_table['Time'][night_2_lims[0]:night_2_lims[1]],i_table['Resid'][night_2_lims[0]:night_2_lims[1]],i_table['Flux_e'][night_2_lims[0]:night_2_lims[1]])
ps.runspec(pmin=20/60, pmax=276.47*2, norm='psd', oversampling=10)

broken_pl_init = models.BrokenPowerLaw1D(amplitude=3.51686205e+01,x_break=1.40269882e-03,alpha_1=0.3564771551659097,alpha_2=1.96878805e+00)
broken_pl_init.x_break.fixed = True
broken_pl_init.alpha_1.fixed = True
fit_pl = fitting.LevMarLSQFitter()

data_table = Table([ps.freq_range,ps.PS], names=('Freq', 'PS'), meta={'name': 'flux_table'})
freq_bin = np.trunc(np.array(data_table['Freq']/(24*60*60)) / 24e-4)
data_table_grouped = data_table.group_by(freq_bin)
data_table_binned = data_table_grouped.groups.aggregate(np.median)
data_table_rms = data_table_grouped.groups.aggregate(np.std)

for i,j in enumerate(np.unique(freq_bin)):
    no_points = len(freq_bin[freq_bin==j])
    data_table_rms['PS'][i] = data_table_rms['PS'][i]/np.sqrt(no_points/oversampling)
    
broken_pl = fit_pl(broken_pl_init, data_table_binned['Freq']/(24*60*60),data_table_binned['PS'], weights=1/data_table_rms['PS'])

plt.figure(figsize=[9,6])
plt.plot(ps.freq_range/(24*60*60),ps.PS,color='grey',alpha=0.4,label='PS')
plt.errorbar(data_table_binned['Freq']/(24*60*60),data_table_binned['PS'],xerr=16e-4,yerr=data_table_rms['PS'],fmt='k.',label='Binned PS')
plt.plot(ps.freq_range/(24*60*60),broken_pl(ps.freq_range/(24*60*60)),color='C0',linewidth=4,label='Broken Powerlaw')
plt.xlim(3e-6,max(ps.freq_range/(24*60*60)))
plt.yscale("log")
plt.xscale("log")
plt.xlabel("Freqency (Hz)",size=14)
plt.ylabel("Power",size=14)
plt.tight_layout()
plt.legend()
plt.show()
```

We next use the simulator object from stingray to simulate 10000 light curves with the same red noise spectrum as the broken powerlaw and the same time sampling right as the data. The power spectra of these light curves are taken and all of them are saved into an array. We also plot two of the fake power spectra (red and green) versus the data to be sure the charateristics are looking reasonable.

```python
# Instantiate simulator object
sim = simulator.Simulator(N=len(i_table['Time'][night_2_lims[0]:night_2_lims[1]]), mean=1,
                          rms=0.1)
# Simulate
test_args_1 = broken_pl.alpha_1.value
test_args_2 = broken_pl.alpha_2.value

simulated_PS = []

for i in np.arange(0,10000,1):
    lc_1 = sim.simulate(test_args_1)
    lc_2 = sim.simulate(test_args_2)

    fake_1 = PowerSpec(i_table['Time'][night_2_lims[0]:night_2_lims[1]], lc_1.counts, i_table['Flux_e'][night_2_lims[0]:night_2_lims[1]])
    fake_1.runspec(pmin=((1/(broken_pl.x_break.value))*u.s).to(u.min).value,pmax=276.47*2,norm='psd', oversampling=10.0)
    fake_2 = PowerSpec(i_table['Time'][night_2_lims[0]:night_2_lims[1]], lc_2.counts, i_table['Flux_e'][night_2_lims[0]:night_2_lims[1]])
    fake_2.runspec(pmin=20/60,pmax=((1/(broken_pl.x_break.value))*u.s).to(u.min).value,norm='psd', oversampling=10.0)
    fake = np.append(fake_1.PS.value,fake_2.PS.value)
    simulated_PS.append(fake)

freq_range = np.append(fake_1.freq_range,fake_2.freq_range)

ps = PowerSpec(i_table['Time'][night_2_lims[0]:night_2_lims[1]], i_table['Resid'][night_2_lims[0]:night_2_lims[1]], i_table['Flux_e'][night_2_lims[0]:night_2_lims[1]])
ps.runspec(pmin=20/60, pmax=276.47*2, norm='psd', oversampling=10)
    
plt.figure(figsize=[12,8])

plt.plot(freq_range/(24*60*60),simulated_PS[0], 'g',alpha=0.5,label='Fake Power Spec')
plt.plot(freq_range/(24*60*60),simulated_PS[99], 'r',alpha=0.5,label='Fake Power Spec')
plt.plot(ps.freq_range/(24*60*60),ps.PS, 'k',alpha=0.5,label='Data')
#plt.xlim(0,500)
plt.yscale("log")
plt.xscale("log")
plt.ylabel("PSD")
plt.xlabel("Cycles / hour")
plt.legend()
plt.show()
```

Finally, we cycle through each frequency bin of the power spectrum, and fit the 10,000 values at each frequency bin with a Gaussian. The 3 sigma limit of this fit is then recorded, and this "3 sigma" limit power spectrum fit with a broken power law to find the amplitdues for plotting in the previous cells (for example, this final cell returns a broken powerlaw with parameter values of amplitude=25107.21553459, x_break=0.0014027, alpha_1=0.35647716, alpha_2=1.85583169. This is what's used back in the big power spectrum plot when plotting the 3 sigma limits on the subplot for i band 2017-10-16).

```python
upperlimits = []

for i in np.arange(0,np.shape(simulated_PS)[1]-300,1):

    lst2 = [item[i] for item in simulated_PS]
    bin_data = np.histogram(np.log10(lst2),50)
    g_init = models.Gaussian1D(amplitude=100, mean=np.median(np.log10(lst2)), stddev=1.0) #+ models.Gaussian1D(amplitude=100, mean=np.median(np.log10(lst2)), stddev=1.0)
    fit_g = fitting.LevMarLSQFitter()
    g = fit_g(g_init, bin_data[1][:-1], bin_data[0])
    
    upperlimits = np.append(upperlimits, np.max([g.mean.value+3*g.stddev.value]))
    
broken_pl_init = models.BrokenPowerLaw1D(amplitude=broken_pl.amplitude.value, x_break=broken_pl.x_break.value, alpha_1=broken_pl.alpha_1.value, alpha_2=broken_pl.alpha_2.value)
broken_pl_init.alpha_1.fixed=True
broken_pl_init.alpha_2.fixed=True
broken_pl_init.x_break.fixed=True
fit_bpl = fitting.LevMarLSQFitter()

broken_pl_ul = fit_bpl(broken_pl_init, freq_range[:-300]/(24*60*60),10**upperlimits)
```
